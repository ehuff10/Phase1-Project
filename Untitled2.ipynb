```python
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm 
```

```python
import pandas as pd
import zipfile
zip_path= "/users/ehuff/Downloads/archive.zip"
csv_filename = "AviationData.csv"
with zipfile.ZipFile(zip_path, 'r') as z:
     df_original = pd.read_csv(z.open(csv_filename), encoding="mac_roman", low_memory=False)
```

```python
with zipfile.ZipFile(zip_path, 'r') as z:
    print(z.namelist())
```

```python
print(df_original.info())  
print(df_original.head())  
```

```python
df_clean = df_original.copy()
```

```python
df_clean['Event.Date'] = pd.to_datetime(df_clean['Event.Date'], errors='coerce')
```

```python
df_clean['Latitude'] = pd.to_numeric(df_clean['Latitude'], errors='coerce')
df_clean['Longitude'] = pd.to_numeric(df_clean['Longitude'], errors='coerce')
```

```python
print(df_clean.isnull().sum().sort_values(ascending=False))
```

```python
df_clean.fillna({
    'Total.Fatal.Injuries': 0,
    'Total.Serious.Injuries': 0,
    'Total.Minor.Injuries': 0,
    'Total.Uninjured': 0,
    'Injury.Severity': 'Unknown',
    'Purpose.of.flight': 'Unknown'
}, inplace=True)
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
df_clean['Year'] = df_clean['Event.Date'].dt.year  # Extract Year
df_clean['Year'].value_counts().sort_index().plot(kind='line', marker='o')

plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.title("Aviation Accidents Over Time")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df_clean['Make'].value_counts().head(10).plot(kind='bar', color='blue')
df_clean['Make'] = df_clean['Make'].str.title() 

plt.xlabel("Aircraft Manufacturer")
plt.ylabel("Number of Accidents")
plt.title("Top 10 Aircraft Manufacturers Involved in Accidents")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df_clean.groupby('Aircraft.Category')['Total.Fatal.Injuries'].mean().sort_values().plot(kind='barh', color='red')

plt.xlabel("Average Fatalities per Accident")
plt.ylabel("Aircraft Category")
plt.title("Fatality Rate by Aircraft Category")
plt.show()
```

```python
import plotly.express as px

fatal_accidents = df_clean[df_clean['Total.Fatal.Injuries'] > 0] 

fig = px.scatter_geo(fatal_accidents, lat="Latitude", lon="Longitude",
                     hover_name="Location", hover_data=["Year", "Aircraft.Category", "Total.Fatal.Injuries"],
                     color_discrete_sequence=["red"], projection="natural earth")

fig.update_layout(title="Global Distribution of Fatal Aviation Accidents")
fig.show()
```

```python
df_clean['Fatal_Accident'] = df_clean['Total.Fatal.Injuries'].apply(lambda x: 'Fatal' if x > 0 else 'Non-Fatal')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=df_clean, x="Fatal_Accident", hue="Fatal_Accident", palette=["blue", "red"], legend=False)

plt.xlabel("Accident Type")
plt.ylabel("Number of Accidents")
plt.title("Comparison of Fatal vs. Non-Fatal Aviation Accidents")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df_clean[df_clean['Fatal_Accident'] == 'Fatal']['Broad.phase.of.flight'].value_counts().plot(kind='bar', color='darkred')

plt.xlabel("Phase of Flight")
plt.ylabel("Number of Fatal Accidents")
plt.title("Fatal Aviation Accidents by Phase of Flight")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
fatal_weather = df_clean[df_clean['Fatal_Accident'] == 'Fatal']['Weather.Condition'].value_counts()
fatal_weather.plot(kind='bar', color='navy')

plt.xlabel("Weather Condition")
plt.ylabel("Number of Fatal Accidents")
plt.title("Impact of Weather on Fatal Accidents")
plt.show()
```

```python
risk_factors = df_clean.groupby('Fatal_Accident').agg({
    'Weather.Condition': lambda x: x.value_counts().idxmax(),  # Most common weather condition
    'Aircraft.Category': lambda x: x.value_counts().idxmax(),  # Most common aircraft type
    'Broad.phase.of.flight': lambda x: x.value_counts().idxmax(),  # Most common phase of flight
    'Total.Fatal.Injuries': 'sum',  # Total fatal injuries
    'Total.Serious.Injuries': 'sum',
    'Total.Minor.Injuries': 'sum',
}).reset_index()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df_clean, x="Broad.phase.of.flight", hue="Fatal_Accident", palette="coolwarm")

plt.xlabel("Phase of Flight")
plt.ylabel("Number of Accidents")
plt.title("Flight Phase and Fatality Rates")
plt.xticks(rotation=45)
plt.legend(title="Accident Type")
plt.show()
```
